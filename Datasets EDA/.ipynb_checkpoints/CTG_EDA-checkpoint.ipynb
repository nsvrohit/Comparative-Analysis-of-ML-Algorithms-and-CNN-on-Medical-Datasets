```python
# Importing Support Libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Loading the Dataset
dataURL = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00193/CTG.xls'
df = pd.read_excel(dataURL, sheet_name = 'Raw Data', header=None)
```

```python
df
```

```python
df = df.dropna()
```

```python
df.columns = df.iloc[0]
```

```python
df = df.drop([0])
```

```python
df = df.drop(['FileName', 'Date', 'SegFile'], axis = 1)
```

```python
df = df.reset_index()
```

```python
df
```

```python
df = df.drop(['index','b', 'e', 'LBE', 'DR', 'A', 'B', 'C', 'D', 'E', 'AD', 'DE', 'LD', 'FS', 'SUSP'], axis = 1)
```

```python
df
```

```python
df.describe()
```

```python
tdf = df
```

```python
tdf = tdf.drop(['CLASS'], axis=1)
```

```python
tdf = tdf.drop(['NSP'], axis=1)
```

```python
tdf.describe()
```

```python
tdf
```

```python
for column in tdf:
    sns.distplot(df[column], hist = True, kde = True, kde_kws = {'shade': True, 'linewidth': 1})
    plt.xlabel(column)
    plt.ylabel('Density')
    plt.show()
```

```python
sns.set(context='poster', style='whitegrid', rc={'figure.figsize':(20,20)})
sns.boxplot(data=tdf, orient='h')
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
tdfs = scaler.fit_transform(tdf)
```

```python
sns.set(context='poster', style='whitegrid', rc={'figure.figsize':(20,20)})
sns.boxplot(data=tdfs, orient='h')
```
