```python
# Importing Support Libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Loading the Dataset
dataURL = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00388/data.csv'
df = pd.read_csv(dataURL, header=[0])
```

```python
df
```

```python
df = df.drop(['Unnamed: 0'], axis=1)
```

```python
for column in df.iloc[:,:179]:
    sns.distplot(df[column], hist = True, kde = True, kde_kws = {'shade': True, 'linewidth': 1})
    plt.xlabel(column)
    plt.ylabel('Density')
    plt.show()
```

```python
plt.figure(figsize=(30,30))
sns.heatmap(df.iloc[:,:178].corr(), cmap = "summer")
```

```python
tdf = df.drop(['y'], axis=1)
```

```python
sns.set(context='poster', style='whitegrid', rc={'figure.figsize':(20,20)})
sns.boxplot(data=tdf, orient='h')
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
tdfs = scaler.fit_transform(tdf)
```

```python
sns.set(context='poster', style='whitegrid', rc={'figure.figsize':(20,20)})
sns.boxplot(data=tdfs, orient='h')
```
